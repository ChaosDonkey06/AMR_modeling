## Model

$$p_r[t+1]=\frac{w_R p_r}{w_Rp_r+w_S p_s}$$

```python
plt.rc('font', size=15) #controls default text size
fig, ax = plt.subplots(1, 1, figsize=(15.5, 7.2))

def model_frequency(pt, a, wr):
    ws = 1-a*(1-pt)
    p_next = wr*pt / (wr*pt+ws*(1-pt))

    return p_next

num_days = 365*2
p0 = 1e-3
ptime = np.full((num_days+1), np.nan)
ptime[0] = p0

aaa = [10, 50, 100, 150]
for a in aaa:
    for t_idx in range(num_days):

        ptime[t_idx+1] = model_frequency(ptime[t_idx],  a=a/1000, wr=0.99)

    ax.plot(ptime, lw=3, label=a)
ax.set_xlabel("Number of days")
ax.legend(title="DDD per 1000 individuals")
ax.set_ylabel("Frequency of resistant strain $p_r$")
ax.set_xlabel("Num days")
```
